# Finding Corelation Between Numeric Columns Through Heatmap

```python
df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```

# Distribution of Numerical Columns

```python
plt.figure(figsize=(8,5))

sns.histplot(
    df['Quantity'],
    kde=True,
    bins=20,
    palette='Set1'
)

plt.title("Distribution of Quantity",fontsize=16,fontweight='bold')

plt.show()
```

**The histogram shows the distribution of product quantities ordered by customers. Most orders contain a small number of items, while larger quantity orders are comparatively less frequent. This indicates that customers generally purchase products in small quantities.**

plt.figure(figsize=(6,4))
sns.histplot(df['UnitPrice'], kde=True)
plt.title("Distribution of Unit Price")
plt.show()

**The distribution of unit prices indicates that most products are priced within a lower to moderate range. Only a few products have significantly higher prices, suggesting the presence of premium-priced items.**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['ItemsInCart'], kde=True)
plt.title("Distribution of Items in Cart")
plt.show()
```

**The majority of customers have a relatively small number of items in their shopping carts. Large cart sizes occur less frequently, indicating that customers typically purchase only a few items per order.**

```python
plt.figure(figsize=(6,4))
sns.histplot(df['TotalPrice'], kde=True)
plt.title("Distribution of Total Price")
plt.show()
```

**Most orders have lower total purchase values, while a smaller number of orders contribute to higher sales amounts. This suggests that high-value purchases are less common than regular purchases.**

# Boxplot For Outliers

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Quantity'])
plt.title("Boxplot of Quantity")
plt.show()
```

**The boxplot helps identify outliers in the quantity ordered. A few unusually large orders are present, but most orders fall within a consistent range.**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['UnitPrice'])
plt.title("Boxplot of Unit Price")
plt.show()
```

**The boxplot reveals several high-priced products that appear as outliers. Most products, however, are priced within a similar range.**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['ItemsInCart'])
plt.title("Boxplot of Items In Cart")
plt.show()
```

**Most customers add a similar number of items to their carts. A few customers have significantly larger cart sizes, which appear as outliers.**

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['TotalPrice'])
plt.title("Boxplot of Total Price")
plt.show()
```

**The boxplot indicates that while most orders have moderate total values, a few orders have exceptionally high total prices, making them outliers.**

# Count Plots For Categorical Columns

```python
plt.figure(figsize=(8,5))

ax = sns.countplot(
    data=df,
    x='PaymentMethod',
    palette='Set2'
)

plt.title("Payment Method Distribution", fontsize=16, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

**The visualization shows that customers have clear preferences for certain payment methods. The most frequently used payment method dominates customer transactions, indicating higher customer trust and convenience.**

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='OrderStatus',
    hue='OrderStatus',
    palette='husl',
    legend=False
)

plt.title("Order Status Distribution",fontsize=16,fontweight='bold')

plt.show()
```

**Most orders fall into successful completion categories such as Delivered or Shipped, while only a small proportion remain pending or cancelled. This indicates efficient order processing.**

```python
plt.figure(figsize=(8,5))

sns.countplot(
    data=df,
    x='ReferralSource',
    hue='ReferralSource',
    palette='Paired',
    legend=False
)

plt.title("Referral Source Distribution",fontsize=16,fontweight='bold')

plt.show()
```

**The chart highlights the effectiveness of different referral channels. One or two referral sources contribute the majority of customers, demonstrating stronger marketing performance through those channels.**

```python
top_products=df['Product'].value_counts().head(10)

plt.figure(figsize=(12,6))

sns.barplot(
    x=top_products.index,
    y=top_products.values,
    hue=top_products.index,
    palette='Spectral',
    legend=False
)

plt.title("Top 10 Best Selling Products",fontsize=16,fontweight='bold')

plt.xticks(rotation=45)

plt.show()
```

**A small group of products accounts for the highest number of sales. These products represent the most popular items and contribute significantly to overall business performance.**

```python
sales=df.groupby('PaymentMethod')['TotalPrice'].sum().reset_index()

plt.figure(figsize=(8,5))

sns.barplot(
    data=sales,
    x='PaymentMethod',
    y='TotalPrice',
    hue='PaymentMethod',
    palette='rocket',
    legend=False
)

plt.title("Total Sales by Payment Method",fontsize=16,fontweight='bold')

plt.show()
```

**Different payment methods contribute varying amounts to total revenue. The payment method generating the highest sales represents the preferred choice for higher-value transactions.**

```python
df['Date']=pd.to_datetime(df['Date'])

monthly_sales=df.groupby(df['Date'].dt.to_period('M'))['TotalPrice'].sum()

plt.figure(figsize=(12,5))

monthly_sales.plot(
    marker='o',
    linewidth=3,
    color='#6C5CE7'
)

plt.title("Monthly Sales Trend",fontsize=16,fontweight='bold')

plt.show()
```

**Monthly sales exhibit fluctuations over time, reflecting seasonal demand and changing customer purchasing behavior. Identifying these trends can help businesses improve inventory planning and promotional strategies.**
